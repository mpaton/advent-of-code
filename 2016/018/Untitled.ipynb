```python
def checkfortraps(l,c,r):
    if l == 2 and c == 2 and r == 1:
        return 2
    if l == 1 and c == 2 and r == 2:
        return 2
    if l == 2 and c == 1 and r == 1:
        return 2
    if l == 1 and c == 1 and r == 2:
        return 2
    return 1
```
