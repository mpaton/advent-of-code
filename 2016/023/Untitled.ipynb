```python
import pandas as pd
import numpy as np
import math
import collections
import re
import sys
import itertools
```

```python
def is_digit(num):
    try:
        int(num)
        is_digit=True
    except ValueError:
        is_digit=False
        
    return is_digit
```

```python
def cpy(cmd,registers):
    value = 0
    if cmd[2] not in registers:
        return
    
    if is_digit(cmd[1]):
        value = int(cmd[1])
    else:
        value = registers[cmd[1]]
    registers[cmd[2]] = value  
    
def sum(cmd,registers):
    if is_digit(cmd[1]):
        value = int(cmd[1])
    else:
        value = registers[cmd[1]]
    registers[cmd[2]] += value
    
def mul(cmd,registers):
    if is_digit(cmd[1]):
        value = int(cmd[1])
    else:
        value = registers[cmd[1]]
    registers[cmd[2]] *= value
    
def jnz(cmd,registers):
    if is_digit(cmd[1]):
        check = int(cmd[1])
    else:
        check = registers[cmd[1]]
        
    if check != 0:
        if is_digit(cmd[2]):
            return int(cmd[2])
        else:
            return registers[cmd[2]]
        
    return 0

def tgl(cmd,idx,commands,registers):
    location = 0
    if cmd[1].isdigit():
        location = idx+int(cmd[1])
    else:
        location = idx+registers[cmd[1]]
    
    if location > len(commands)-1:
        return
    
    tglcmd = commands[location]
    
    if len(tglcmd) == 2:
        if tglcmd[0] == 'inc':
            tglcmd[0] = 'dec'
        else:
            tglcmd[0] = 'inc'
    else:
        if tglcmd[0] == 'jnz':
            tglcmd[0] = 'cpy'
        else:
            tglcmd[0] = 'jnz'
            
    commands[location] = tglcmd
```
