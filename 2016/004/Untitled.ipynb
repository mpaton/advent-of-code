```python
import pandas as pd
import numpy as np
import math
import collections
```

```python
data = pd.read_csv('./rooms.txt',header=None,delim_whitespace=True)
```

```python
def buildchecksum(name):
    commonchar = collections.Counter(name).most_common(26)
    
    checksum = ''
    idx = 0
    while idx < len(commonchar):
        elem = commonchar[idx]
        ties = [i for i, x in enumerate(commonchar) if x[1] == elem[1]]
        if len(ties) == 1:
            checksum += elem[0]
            idx += 1
        else:
            sorted_ties = ''
            ### we need to sort
            for tie in ties:
                sorted_ties += commonchar[tie][0]
            checksum += ''.join(sorted(sorted_ties))
            idx = ties[-1]+1
    return checksum[0:5]
```

```python
def decrypt(name,sectorid):
    shiftby = divmod(sectorid,26)[1]
    print shiftby
    room = ''
    for elem in name:
        decrypt = ord(elem) + shiftby
        if decrypt > 122:
            decrypt -= 26
        room += chr(decrypt)
    return room 
```

```python
valid_sum = 0
for room in data[0]:
    splitroom = room.split('-')
    ## partition the data
    name = splitroom[:-1]
    name = ''.join(name)
    sectorid = int(splitroom[-1:][0][0:3])
    checksum = splitroom[-1:][0][4:9]
    
    if buildchecksum(name) == checksum:
        valid_sum += sectorid
        ## decrypt
        print 'VALID: ' + decrypt(name,sectorid), sectorid
        
    else:
        ## decrypt
        print 'INVALID: ' + decrypt(name,sectorid), sectorid
print valid_sum
```
