```python
import pandas as pd
import numpy as np
import math 
```

```python
plan = pd.read_csv('./plan.csv', header=None)
```

```python
def Rotate(turn,theta):
    if(turn == 'L'):
        dtheta = -math.pi/2
    elif(turn == 'R'):
        dtheta = math.pi/2
    theta += dtheta
    
    if(theta < 0):
        theta += math.pi*2
    if(theta > math.pi*2):
        theta -= math.pi*2
    return theta
```

```python
x = 0
y = 0
theta = 0

blocks = 0
previous_orientation = ''
for idx,value in plan.iteritems():
    directions_str = value[0]
    rot = directions_str[0]
    trans = np.matrix([[float(directions_str[1:])],[0]])
    ## Rotate by the turn amount.
    theta = Rotate(rot,theta)
    rot_mat = np.matrix([[math.cos(theta), -math.sin(theta)],
                         [math.sin(theta), math.cos(theta)]])

    trans = rot_mat * trans
    x = x + trans[0]
    y = y + trans[1]
    print x,y
print x,y
print math.fabs(x) + math.fabs(y)
```
